# Pandas Visualization 

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
import pandas as pd
```

```python
num_samples = 100

ts = pd.Series(
    np.random.randn(num_samples),
    index=pd.date_range(
        "1/1/2000",
        periods=num_samples)
)
```

```python
print(ts.head())
```

```python
ts.plot();
```

```python
num_samples = 100
num_features = 2

df = pd.DataFrame(
    np.random.randn(num_samples, num_features),
    index=ts.index,
    columns=["A", "B"]
)
```

```python
print(df.head())
```

```python
df.plot();
```

Plotting methods allow for a handful of plot styles other than the default line plot. These methods can be provided as the kind keyword argument to plot(), and include:

- ‘bar’ or ‘barh’ for bar plots
- ‘hist’ for histogram
- ‘box’ for boxplot
- ‘kde’ or ‘density’ for density plots
- ‘area’ for area plots
- ‘scatter’ for scatter plots
- ‘hexbin’ for hexagonal bin plots
- ‘pie’ for pie plots

```python
df.plot(kind="hist");
```

```python
df.plot(kind="scatter", x="A", y="B");
```
