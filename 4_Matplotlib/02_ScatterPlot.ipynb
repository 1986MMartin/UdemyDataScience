# Scatter Plots

````python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
np.random.seed(0)
````

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
````

````python
plt.plot(x, y, '-o');
````

````python
plt.plot(
    x, y,
    '--',
    color='gray',
    markersize=15,
    linewidth=4,
)
plt.ylim(-1.2, 1.2);
````

```python
matplotlib.pyplot.scatter(
    x, y,
    s=None,
    c=None,
    marker=None,
    ...,
    **kwargs
)
```

Marker:

- 'v'
- '^'

- 'd'

````python
plt.scatter(x, y, marker='o');
````

````python
plt.scatter(x, y, marker='X');
````

````python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)

plt.scatter(
    x, y,
    c=colors,
    s=10,
    alpha=0.3,
    cmap='viridis'
)
plt.colorbar();
````
