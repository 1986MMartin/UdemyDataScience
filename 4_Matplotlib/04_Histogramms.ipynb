# Histograms

````python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
np.random.seed(0)
````

````python
data = np.random.randn(1000)
````

```python
plt.hist(
    x,
    bins=None,
    range=None,
    density=False,
    weights=None,
    cumulative=False, 
    color=None, 
    label=None,
    ...,
    **kwargs
)
```

````python
plt.hist(data);
````

````python
plt.hist(
    data,
    bins=30,
    alpha=0.5,
    histtype='stepfilled',
    color='steelblue',
);
````

````python
counts, bin_edges = np.histogram(data, bins=5)

print(counts)
````

````python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(
    histtype='stepfilled',
    alpha=0.3,
    bins=40
)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
````

## Two-Dimensional Histograms and Binnings

````python
mean = [0, 0]
cov = [[1, 1], [1, 2]]

x, y = np.random.multivariate_normal(
    mean=mean,
    cov=cov,
    size=10000
).T
````

````python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar();
````
